## ‣ Problem 1: Real-World Everyday-Use Problem 

**Problem Statement:**
You're working on a dashboard to visualize employee participation in a company project. There are 5 developers who worked for different durations (in hours) on the project. The total duration allotted for the project was 8 hours. Create a visualization to show how much each developer contributed out of the total time, and make the chart:

- Visually clear and readable

- Emphasize the highest contributing developer

- Show each portion as a percentage

**Input ➤**

{
  "Vishant": 2,
  "Manpreet": 1,
  "Pramod": 2.5,
  "Riya": 1,
  "Sneha": 1.5
}

**Output ➤**

- A clean, readable chart that displays each person’s contribution as a percentage of total time.

- The largest contributor should stand out clearly from the rest.

**Code ➤**

```python
# importing
from matplotlib import pyplot as plt 

plt.style.use('fivethirtyeight')

def project_contribution(data):
    # Data cleaning for graph building
    largest=max(data, key=data.get)
    workers=[]
    hours=[]
    explode=[]
    for i in data:
        workers.append(i), hours.append(data[i])
        if(i==largest):
            explode.append(0.2)
        else:
            explode.append(0)
        
    
    # Making Pie 
    plt.pie(hours,labels=workers,autopct="%1.1f%%", explode=explode,startangle=90)
    
    plt.title("Project Contribution")
    plt.tight_layout()
    return plt.gcf()

data= { "Vishant": 2, "Manpreet": 1, "Pramod": 2.5, "Riya": 1, "Sneha": 1.5 }
fig= project_contribution(data) 
plt.show()
```

## ‣ Problem 2: Difficult Problem 

**Problem Statement:**
You are given daily temperature data (in °C) for two cities—City A and City B—over 10 days. Your task is to compare their temperature trends visually. Plot a line chart for both cities and use a shaded area to indicate where:

- City A was warmer than City B — shaded red

- City B was warmer than City A — shaded blue

- Make sure the plot is informative and visually clean, with proper labels, legends, and shaded regions without clipping at intersections.



**Input ➤**

```python
days = list(range(1, 11))
city_a = [30, 32, 33, 31, 29, 28, 27, 30, 32, 31]
city_b = [28, 31, 34, 32, 30, 30, 29, 28, 30, 32]
```

**Output ➤**

A dual-line plot comparing both cities with:

- Red-filled area where City A > City B

- Blue-filled area where City B > City A

- Custom line thickness, legend positioning, and no intersection clipping

**Code ➤**

```python
import numpy as np

def compare_temperature(days,city_a,city_b):
    #plotting and comparison 
    plt.plot(days, city_a,label='City A ',linewidth=1.5,marker='>')
    
    plt.plot(days,city_b,label="City B ",linewidth=1.5,marker='o')
    
    city_a = np.array(city_a)
    city_b = np.array(city_b)
    
    plt.fill_between(days,city_a,city_b,where=(city_a > city_b),interpolate=True,color="red",alpha=0.5,label="City A > City B")
    plt.fill_between(days,city_a,city_b,where=(city_a < city_b),interpolate=True,color='blue',alpha=0.5,label="City A < City B")
    
    plt.title("Temperature for City A and B")
    plt.xlabel("Days")
    plt.ylabel("Temperatures")
    plt.legend(loc=(1,1))
    
    
    return plt.gcf()

#input
days = list(range(1, 11))
city_a = [30, 32, 33, 31, 29, 28, 27, 30, 32, 31]
city_b = [28, 31, 34, 32, 30, 30, 29, 28, 30, 32]

fig= compare_temperature(days,city_a,city_b)
plt.show()
```
